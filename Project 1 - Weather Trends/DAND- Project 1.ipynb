```python
def compute_error(b, m, data):
    Error = 0
    for i in range(0, len(data)):
        x = data[i, 0]/max(temp_data['year'])
        y = data[i,1]/max(temp_data['global_data'])
        Error += (y - (m*x + b))**2
    return Error/float(len(data))

def step_gradient(b,m,data,learning_rate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(data))
    for i in range(0, len(data)):
        x = data[i, 0]/2015
        y = data[i, 1]/9.73
        b_gradient += -(2/N)*(y-(m*x + b))
        m_gradient += -(2/N)*x*(y-(m*x + b))
    new_b = b - (learning_rate*b_gradient)
    new_m = m - (learning_rate*m_gradient)  
    return [new_b, new_m]

def gradient_descent_runner(data, learning_rate, b_start, m_start, iterations):
    b = b_start
    m = m_start
    for i in range(iterations):
        b, m = step_gradient(b, m, array(data), learning_rate)
    return [b,m]
    
def run():
    data = genfromtxt('global_temp_data.csv', delimiter=',', skip_header=1)
    print(data)
    learning_rate = 0.0001
    initial_b = 0
    initial_m = 0 
    num_iter = 5000
    [b,m] = gradient_descent_runner(data, learning_rate, initial_b, initial_m, num_iter)
    print(b)
    print(m)
    
if __name__ == '__main__':
    run()
```
